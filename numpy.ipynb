```python
kanto = [73,67,43]
johto = [91,88,64]
hoenn = [87,134,58]
sinnoh = [102,43,37]
unova = [69,96,70]
```

```python
weights = [w1,w2,w3]
```

```python
%%time
def crop_yield(region,weights):
  result = 0
  for x,w in zip(region,weights):
    result += x*w
  # for i in range(len(region)):
  #   result += region[i]*weights[i]
  return result
```

```python
crop_yield(kanto,weights)
```
